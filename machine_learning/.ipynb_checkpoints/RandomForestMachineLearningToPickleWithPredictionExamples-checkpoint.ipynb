#### Changing Type to match data 

```python
merged_data.dtypes
```

```python
merged_data['day_of_week_x'] = merged_data['day_of_week_x'].astype('category')

merged_data['hours_x'] = merged_data['hours_x'].astype('category')

merged_data['main'] = merged_data['main'].astype('category')
merged_data['description'] = merged_data['description'].astype('category')

```

```python
merged_data.dtypes
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(merged_data['day_of_week_x'])
# Drop column B as it is now encoded
merged_data = merged_data.drop('day_of_week_x',axis = 1)
# Join the encoded df
merged_data = merged_data.join(one_hot)
merged_data
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(merged_data['hours_x'])
# Drop column B as it is now encoded
merged_data = merged_data.drop('hours_x',axis = 1)
# Join the encoded df
merged_data = merged_data.join(one_hot)
merged_data
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(merged_data['main'])
# Drop column B as it is now encoded
merged_data = merged_data.drop('main',axis = 1)
# Join the encoded df
merged_data = merged_data.join(one_hot)
merged_data
```

```python
# Get one hot encoding of columns B
one_hot = pd.get_dummies(merged_data['description'])
# Drop column B as it is now encoded
merged_data = merged_data.drop('description',axis = 1)
# Join the encoded df
merged_data = merged_data.join(one_hot)
merged_data
```

```python
merged_data.dtypes
```

```python
merged_data = merged_data.reindex(columns = [col for col in merged_data.columns if col != 'available_bikes'] + ['available_bikes'])
merged_data
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.2, random_state=123)
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model on the training data
rf.fit(X_train, y_train)

# Predict bike availability on the testing data
y_pred = rf.predict(X_test)

# Evaluate the model performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean squared error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
```
